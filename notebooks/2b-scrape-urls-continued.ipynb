```python
import psycopg2
conn = psycopg2.connect("host=localhost dbname=postgres user=postgres")
cur = conn.cursor()

table_name = "bq_all"

cur.execute("SELECT id, url FROM {} WHERE article_content IS NULL OR article_content = '';".format(table_name))
urls_list = cur.fetchall()
```

```python
len(urls_list)
```

```python
from newspaper import Article

done_indices = []

# Write function for getting article text and summary:
def update_article_content(table_name, field, identifier, data_str):
    conn = psycopg2.connect("host=localhost dbname=postgres user=postgres")
    cur = conn.cursor()
    SQL = "UPDATE {} SET {}=(%s) WHERE id={};".format(table_name, field, identifier)
    data = (data_str, )
    cur.execute(SQL, data)
    conn.commit()
    return 


def get_article(article_obj):
    article_obj.download()
    article_obj.parse()
    article_obj.nlp()
    article_content = article_obj.text
    article_summary = article_obj.summary
    return article_content, article_summary

def get_text_and_summary(index):  
    identifier = urls_list[index][0]
    url = urls_list[index][1]
    
    if type(url) == str and url != "":
        article = Article(url)
        try:
            article_content, article_summary = get_article(article)
            if type(article_content) == str:
                update_article_content(table_name, "article_content", identifier, article_content)
            else:
                update_article_content(table_name, "article_content", identifier, '')
            if type(article_summary) == str:
                update_article_content(table_name, "article_summary", identifier, article_summary)
            else:
                update_article_content(table_name, "article_summary", identifier, '')
        except:
            update_article_content(table_name, "article_content", identifier, '')
            update_article_content(table_name, "article_summary", identifier, '')
    else:
        update_article_content(table_name, "article_content", identifier, '')
        update_article_content(table_name, "article_summary", identifier, '')
    
    done_indices.append(index)
    
    return
```

```python
%%time

from multiprocessing import Pool

p = Pool()  
records = p.map(get_text_and_summary, range(len(urls_list)))
p.terminate()
p.join()
```
