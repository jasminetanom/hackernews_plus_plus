## Query PostgreSQL DB for URLs

```python
import psycopg2
conn = psycopg2.connect("host=localhost dbname=postgres user=postgres")
cur = conn.cursor()

table_name = "bq_all"

cur.execute("ALTER TABLE {} ADD COLUMN article_content TEXT;".format(table_name))
cur.execute("ALTER TABLE {} ADD COLUMN article_summary TEXT;".format(table_name))

conn.commit()
```

```python
cur.execute("SELECT id, url FROM {};".format(table_name))
urls_list = cur.fetchall()
```

```python
urls_list[:5]
```

## Scrape article text and content from URLs

```python
from newspaper import Article

# Write function for getting article text and summary:
def get_text_and_summary(index):
    conn = psycopg2.connect("host=localhost dbname=postgres user=postgres")
    cur = conn.cursor()
    
    identifier = urls_list[index][0]
    url = urls_list[index][1]
    
    if type(url) == str and url != "":
        article = Article(url)
        try:
            article.download()
            article.parse()
            article.nlp()
            article_content = article.text
            article_summary = article.summary
            if type(article_content) == str:
                
                SQL = "UPDATE {} SET article_content=(%s) WHERE id={};".format(table_name, identifier)
                data = (article_content, )
                cur.execute(SQL, data)
            if type(article_summary) == str:
                SQL = "UPDATE {} SET article_summary=(%s) WHERE id={};".format(table_name, identifier)
                data = (article_summary, )
                cur.execute(SQL, data)
        except:
            pass
    if index % 100 == 0:
        print("Completed scraping of {} articles".format(index))
    
    conn.commit()
    
    return
```

```python
from multiprocessing import Pool

p = Pool(10)  
pool = multiprocessing.Semaphore(multiprocessing.cpu_count())
records = p.map(get_text_and_summary, range(len(urls_list)))
p.terminate()
p.join()
```
